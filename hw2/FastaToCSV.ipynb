```python
import numpy as np
import pandas as pd
import csv
```

```python
import Bio
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
HG = list(SeqIO.parse('HW2_reads.fasta', "fasta"))
```

```python
HG
```

```python
df = pd.DataFrame()
ID = []
Seq = []

for r in HG: 
    ID.append(r.name)
    Seq.append(r.seq)
    
df['ID'] = ID
df['Sequence'] = Seq 
```
